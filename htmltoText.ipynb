```python
import re
import sys
class html2Text:
    def __init__(self):
        self.state = 0
        self.s = ""
    def htm2Text(self, html):
        self.s = ""
        for ch in html:
            if self.state == '<':
                if ch != '>':
                    continue
                self.state = 0
            else:
                if ch == '<':
                    self.state = '<'
                    continue
                self.s += ch
        return self.s.strip()
    def read1(self, filename):
        with open(filename, 'r') as sentences:
            f = sentences.read()
        fin_string = ""
        for line in f:
            fin_string += line
        return fin_string.strip()
    def main(self):
        # s = self.htm2Text("<p>Hello string</p>")
        # print (s)
        s1 = self.read1("site.html")
        s2 = self.htm2Text(s1)
        print (s2)

if __name__ == "__main__":
    html2Text().main()
```
